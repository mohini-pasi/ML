```python
import numpy as np 
import pandas as  pd 
import matplotlib.pyplot as plt
import seaborn as sns


import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\Rohit\Downloads\cars.csv")
```

```python
df
```

```python
df["normalized-losses"]= df["normalized-losses"].replace("?",np.nan)
```

```python
df["horsepower"]= df["horsepower"].replace("?",np.nan)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
warnings.filterwarnings("ignore")
```

```python
df["normalized-losses"]= df["normalized-losses"].astype(float)
df["horsepower"]= df["horsepower"].astype(float)
```

```python
df["normalized-losses"].fillna( df["normalized-losses"].mean(),inplace=True)
df["horsepower"].fillna( df["horsepower"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.columns
```

```python
catcol=df.select_dtypes(object)
```

```python
catcol.head(3)
```

```python
df['make'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in catcol:
    df[i]=le.fit_transform(df[i])
```

```python
df[i]
```

```python
df.head(3)
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df["price"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
print(r2_score(ytest,ypred))
```

```python
train=lr.score(xtrain,ytrain)
test=lr.score(xtest,ytest)
print(f"Training Result: {train}")
print(f"Testing Result: {test}")
```

training acc high
testing acc low

high bais and low varience --- overfitting

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
l2 = Ridge(alpha = 5)
```

```python
l2.fit(xtrain,ytrain)
```

```python
ypred=l2.predict(xtest)
```

```python
print(r2_score(ytest,ypred))
```

```python
train=l2.score(xtrain,ytrain)
test=l2.score(xtest,ytest)
print(f"Training Result: {train}")
print(f"Testing Result: {test}")
```

```python
for i in range(1,30):
    l2=Ridge(alpha=i)
    l2.fit(xtrain,ytrain)
    test=l2.score(xtest,ytest)
    print(f"The Value of Lambda {i} test score {test}")
```

```python
l2 = Ridge(alpha = 11)
l2.fit(xtrain,ytrain)
ypred=l2.predict(xtest)
print(r2_score(ytest,ypred))
```

```python
train=l2.score(xtrain,ytrain)
test=l2.score(xtest,ytest)
print(f"Training Result: {train}")
print(f"Testing Result: {test}")
```
